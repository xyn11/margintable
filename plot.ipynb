```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
records = pd.read_csv('output.csv')
```

```python
df = pd.DataFrame(records)
```

```python
df['Month/Year'] = df['Month'].map(str) + '/' + df['Year'].map(str)
```

```python
df['SumCredit'] = df["Free Credit Balances in Customers' Cash Accounts"] + df["Free Credit Balances in Customers' Securities Margin Accounts"]
```

```python
df.head(5).transpose()
```

```python
ax = df.plot(x = 'Month/Year', y = ["Debit Balances in Customers' Securities Margin Accounts", "Free Credit Balances in Customers' Cash Accounts",\
       "Free Credit Balances in Customers' Securities Margin Accounts", "SumCredit"], style = ['ro', 'g^', 'b+', 'y*'], figsize = (25, 7))
ax.set_ylabel('in millions')
ax.set_title('Margin Statistics')
```

```python
fig = ax.get_figure()
fig.savefig('plot.png')
```
